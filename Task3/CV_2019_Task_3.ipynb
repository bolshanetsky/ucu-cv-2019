```
import cv2
import numpy as np
import imutils
from skimage import io
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow


def url_to_image(url):
#   print("downloading %s" % (url))
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)

url_1 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj1.jpg"
url_2 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj2.jpg"
url_3 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj3.jpg"
url_4 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj4.jpg"
url_5 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj5.jpg"
url_6 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj6.jpg"

urls = [url_1, url_2, url_3, url_4, url_5, url_6]
```

```
#image 1
img = url_to_image(url_1)
# img = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
# gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blue_channel = img[..., 0]

# print(img.shape)
(w,h) = img.shape[:2]
thresh = cv2.threshold(blue_channel, 200, 255, cv2.THRESH_BINARY)[1]
mask = np.zeros( (w,h),np.uint8)
# mask[thresh == 0] = 3
# mask[thresh == 255] = 1
# print(mask.shape)
# print(mask_test.shape)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect = (39,25,450, 239)
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,2,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
cv2_imshow(mask2*255)
cv2_imshow(img)
```
